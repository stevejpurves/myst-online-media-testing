```python
#| label: my-image

import requests
from IPython.display import Image
url = "https://github.com/Octoframes/jupyter_capture_output/blob/main/assets/cute_dog.jpg?raw=true"
response = requests.get(url)
Image(response.content , embed=True, width=400)
```
